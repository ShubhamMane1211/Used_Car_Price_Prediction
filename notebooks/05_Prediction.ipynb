```python
import pandas as pd
import numpy as np
import joblib
import warnings

warnings.filterwarnings("ignore")
```

```python
model = joblib.load("../models/best_model.pkl")

features = joblib.load("../models/features.pkl")

make_encoder = joblib.load("../models/make_encoder.pkl")
```

```python
print(features)
```

```python
new_car = {

    'symboling':1,

    'normalized-losses':120,

    'fuel-type':'gas',

    'aspiration':'std',

    'num-of-doors':'four',

    'body-style':'sedan',

    'drive-wheels':'fwd',

    'engine-location':'front',

    'wheel-base':97,

    'length':170,

    'width':65,

    'height':54,

    'curb-weight':2400,

    'engine-type':'ohc',

    'num-of-cylinders':'four',

    'engine-size':130,

    'fuel-system':'mpfi',

    'bore':3.46,

    'stroke':3.19,

    'compression-ratio':9,

    'horsepower':111,

    'peak-rpm':5000,

    'city-mpg':24,

    'highway-mpg':30,

    'make':'toyota'
}

sample = pd.DataFrame([new_car])
```

```python
sample['make'] = sample['make'].map(make_encoder)
```

```python
sample = pd.get_dummies(sample)
```

```python
sample = sample.reindex(columns=features, fill_value=0)
```

```python
prediction = model.predict(sample)
```

```python
print("="*40)

print(f"Estimated Car Price : ₹ {prediction[0]:,.2f}")

print("="*40)
```
